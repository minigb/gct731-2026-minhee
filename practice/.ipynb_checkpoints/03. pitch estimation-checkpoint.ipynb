```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np

import tensorflow as tf
import crepe
import pesto

import torch
import torchaudio
from rmvpe import RMVPE


device = "cuda" if torch.cuda.is_available() else "cpu"
model_rmvpe = RMVPE("rmvpe.pt", is_half=False, device=device)
```

# Monophonic Vocal 

```python
audio1 = "../audio/Heyyeahyeah.wav"
y1, sr = librosa.load(audio1, sr=16000)

print(sr)

# PYIN
f0_pyin, voiced_flag, voiced_probs = librosa.pyin(y1, sr=16000, fmin=librosa.note_to_hz('C3'),fmax=librosa.note_to_hz('C6'), hop_length=1024)
times = librosa.times_like(f0_pyin, sr=sr)

print(f0_pyin.shape)

# Crepe
time, f0_crepe, confidence, activation = crepe.predict(y1, sr, step_size=10,  viterbi=True)
times = librosa.times_like(f0_crepe, sr=sr)

print(f0_crepe.shape)


# PESTO
time, f0_pesto, confidence, activation = pesto.predict(y1, sr, step_size=10,  viterbi=True)
times = librosa.times_like(f0_pesto, sr=sr)

print(f0_pesto.shape)

# RMVPE
y1_torch, sr = torchaudio.load(audio1)
if sr != 16000:
    resampler = torchaudio.transforms.Resample(sr, 16000)
    y1_rmvpe = resampler(y1_torch)

if y1_rmvpe.shape[0] > 1:
    y1_rmvpe = y1_rmvpe.mean(dim=0)
else:
    y1_rmvpe = y1_rmvpe.squeeze()    
    
f0_rmvpe = model_rmvpe.infer_from_audio(y1_rmvpe, thred=0.03)
times = librosa.times_like(f0_rmvpe, sr=sr)

print(f0_rmvpe.shape)

if torch.is_tensor(f0_rmvpe):
    print('tensor')
    f0_rmvpe_np = f0_rmvpe.detach().cpu().numpy()

#f0_rmvpe = np.array(f0_rmvpe)    
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import torch
import crepe
import pesto

# --- 1. 기본 설정 ---
SR = 16000
STEP_SIZE_MS = 10
HOP_LENGTH = 160  # 16000Hz * 0.01s
audio_path = "../audio/Heyyeahyeah.wav"
y, _ = librosa.load(audio_path, sr=SR)

# --- 2. Pitch Estimation & 시간축 보정 ---

# (1) PYIN
f0_pyin, voiced_flag, _ = librosa.pyin(y, sr=SR, fmin=100, fmax=1000, hop_length=HOP_LENGTH)
t_pyin = librosa.times_like(f0_pyin, sr=SR, hop_length=HOP_LENGTH) # 이미 초 단위

# (2) CREPE
t_crepe, f0_crepe, conf_crepe, _ = crepe.predict(y, SR, step_size=STEP_SIZE_MS, viterbi=True, verbose=False)
# [보정] 만약 t_crepe가 ms 단위라면 1000으로 나눔
if t_crepe[-1] > len(y)/SR * 10: # 전체 재생시간보다 훨씬 크다면
    t_crepe = t_crepe / 1000.0

# (3) PESTO
y_torch = torch.from_numpy(y).float().unsqueeze(0)
t_pesto, f0_pesto, conf_pesto, _ = pesto.predict(y_torch, SR, step_size=STEP_SIZE_MS)

# [보정] PESTO 타임축 초 단위로 강제 변환
#t_pesto = np.array(t_pesto).flatten()
#if t_pesto[-1] > len(y)/SR * 10:
t_pesto = t_pesto / 1000.0

f0_pesto = np.array(f0_pesto).flatten()
conf_pesto = np.array(conf_pesto).flatten()

# (4) RMVPE
try:
    f0_rmvpe = model_rmvpe.infer_from_audio(y, thred=0.03)
    f0_rmvpe = f0_rmvpe.detach().cpu().numpy().flatten() if torch.is_tensor(f0_rmvpe) else np.array(f0_rmvpe).flatten()
    t_rmvpe = librosa.times_like(f0_rmvpe, sr=SR, hop_length=HOP_LENGTH)
except:
    f0_rmvpe = np.array([])

# --- 3. 시각화 (시간축 정렬 완료) ---

plt.figure(figsize=(15, 7))

# 스펙트로그램 배경 (x_axis='time'은 초 단위를 기대함)
D = librosa.amplitude_to_db(np.abs(librosa.stft(y, n_fft=2048, hop_length=HOP_LENGTH)), ref=np.max)

#img = librosa.display.specshow(D1, x_axis='time', y_axis='log', ax=ax)
img = librosa.display.specshow(D, sr=SR, hop_length=HOP_LENGTH, x_axis='time', y_axis='log', cmap='magma', alpha=0.6)

# 마스킹 및 플로팅 함수
def plot_pitch(t, f, conf, label, color, thresh=0.3, style='-'):
    if f.size == 0: return
    # 신뢰도가 낮은 구간은 마스킹
    f_masked = np.copy(f)
    f_masked[conf < thresh] = 0
    
    # 0이 아닌 값만 골라서 플롯 (선 끊김 방지하려면 슬라이싱 필요)
    mask = f_masked > 0
    plt.plot(t[mask], f_masked[mask], label=label, color=color, linestyle=style, linewidth=2)

# 각 모델별 플롯 (PYIN은 voiced_flag 사용)
plt.plot(t_pyin[voiced_flag], f0_pyin[voiced_flag], label='PYIN', color='cyan', linewidth=2)
plot_pitch(t_crepe, f0_crepe, conf_crepe, 'CREPE', 'lawngreen', thresh=0.5)
plot_pitch(t_pesto, f0_pesto, conf_pesto, 'PESTO', 'white', thresh=0.35, style='--')

if f0_rmvpe.size > 0:
    # RMVPE는 내부적으로 마스킹이 되어 나오므로 직접 플롯
    plt.plot(t_rmvpe[f0_rmvpe > 0], f0_rmvpe[f0_rmvpe > 0], label='RMVPE', color='orange', linewidth=1.5)

plt.ylim(100, 1000)
plt.xlim(0, len(y)/SR) # 시간축 범위를 실제 오디오 길이에 맞게 강제 제한
plt.title("Pitch Comparison (Time Axis Corrected)")
plt.xlabel("Time (s)")
plt.legend(loc='upper right')
plt.show()
```
